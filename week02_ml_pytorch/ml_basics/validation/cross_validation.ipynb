Classifier for predicting likelihood of a passenger on the titanic surviving, based on factors such as name, age, gender, class.

```python
#import required modules and read csv

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)

df = pd.read_csv('titanic_dataset.csv')
```

```python
#define features and label 
y = df['Survived']
x = df[["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]].copy()

# Fill missing numeric/categorical values
x["Age"] = x["Age"].fillna(x["Age"].median())
x["Fare"] = x["Fare"].fillna(x["Fare"].median())
x["Embarked"] = x["Embarked"].fillna(x["Embarked"].mode()[0])

# Convert category text to numeric columns
x = pd.get_dummies(x, columns=["Sex", "Embarked"], drop_first=True)

```

```python
#create train/test split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=1)

```

```python
#create and train regression models
model1 = LogisticRegression(max_iter=1000)

model1.fit(x_train, y_train)

pred1 = model1.predict(x_test)

scores = cross_val_score(
    model1,
    x,
    y,
    cv=5
)

print(scores)
print(scores.mean())
```

```python
#get metrics
accuracy1 = accuracy_score(y_test, pred1)
precision1 = precision_score(y_test, pred1)
recall1 = recall_score(y_test, pred1)
f1_score1 = f1_score(y_test, pred1)

print(accuracy1, precision1, recall1, f1_score1)
```
