Classifier for predicting likelihood of a passenger on the titanic surviving, based on factors such as name, age, gender, class.

```python
#import required modules and read csv

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)

df = pd.read_csv('titanic_dataset.csv')
```

```python
#define features and label 
y = df['Survived']
x = df[["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]].copy()

# Fill missing numeric/categorical values
x["Age"] = x["Age"].fillna(x["Age"].median())
x["Fare"] = x["Fare"].fillna(x["Fare"].median())
x["Embarked"] = x["Embarked"].fillna(x["Embarked"].mode()[0])

# Convert category text to numeric columns
x = pd.get_dummies(x, columns=["Sex", "Embarked"], drop_first=True)

```

```python
#create train/test split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=1)
```

```python
#create and train regression models
model1 = LogisticRegression()
model2 = DecisionTreeClassifier()

model1.fit(x_train, y_train)
model2.fit(x_train, y_train)

pred1 = model1.predict(x_test)
pred2 = model2.predict(x_test)
```

```python
#get metrics
accuracy1 = accuracy_score(y_test, pred1)
precision1 = precision_score(y_test, pred1)
recall1 = recall_score(y_test, pred1)
f1_score1 = f1_score(y_test, pred1)

accuracy2 = accuracy_score(y_test, pred2)
precision2 = precision_score(y_test, pred2)
recall2 = recall_score(y_test, pred2)
f1_score2 = f1_score(y_test, pred2)
```

```python
#display metrics in a table 
results = pd.DataFrame({
    "Logistic Regression": [
        accuracy1,
        precision1,
        recall1,
        f1_score1
    ],
    "Decision Tree": [
        accuracy2,
        precision2,
        recall2,
        f1_score2
    ]
},
index=["Accuracy", "Precision", "Recall", "F1-score"])

print(results)

#export to latex table
table = results.to_latex()
print(table)
```
